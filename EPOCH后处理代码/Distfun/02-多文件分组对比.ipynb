```python
import os
import numpy as np
import matplotlib.pyplot as plt
import sdf_helper as sh  # 自定义的sdf数据读取模块，确保可用

# === 参数设置 ===
E_J_to_MeV = 1.0 / 1.6e-13  # 焦耳转换成MeV的系数
step = 2                    # 合并bin的大小（多少个bin合并成一个）
# species = "Electron"         # 粒子种类，可以改为 "Photon" 等
species = "Photon"         # 粒子种类，可以改为 "Photon" 等
save_fig = False             # 是否保存图片的开关，True保存，False不保存
output_dir = "./figures"     # 保存图片的目录路径
group_size = 5  # 每组绘图的文件数量

# === 多个数据集配置 ===
datasets = [
    {
        'base_dir': '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024a/10kev无磁场',
        'file_nums': range(20, 35),  # 文件编号范围
        'file_prefix': 'distfun',
        'file_suffix': '.sdf',
    },
    {
        'base_dir': '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024ab+2/10kev25磁场',
        'file_nums': range(20, 35),
        'file_prefix': 'distfun',
        'file_suffix': '.sdf',
        # 'custom_labels':[],
    },
]

# === 创建保存目录（如果需要保存且目录不存在则创建）===
if save_fig and not os.path.exists(output_dir):
    os.makedirs(output_dir)

# === 生成完整文件路径和对应标签 ===
file_paths = []  # 用来存放所有文件的完整路径
labels = []      # 用来存放对应的标签（默认用文件名）

for dataset in datasets:
    base_dir = dataset['base_dir']
    prefix = dataset.get('file_prefix', '')  # 获取文件名前缀，默认空字符串
    suffix = dataset.get('file_suffix', '')  # 获取文件名后缀，默认空字符串
    nums = dataset['file_nums']               # 文件编号列表或range
    custom_labels = dataset.get('custom_labels', None)  # 自定义标签（可选）

    for idx, num in enumerate(nums):  # idx当前编号在列表里的索引（位置）,num则是编号比如15、16、17
        # 构建完整文件名
        fname = f"{prefix}{num:04d}{suffix}"  # 例如 distfun0015.sdf
        fpath = os.path.join(base_dir, fname) # 生成完整路径
        file_paths.append(fpath)

        if custom_labels:
            labels.append(custom_labels[idx])   # 如果有自定义标签，使用自定义标签
        else:
            labels.append(os.path.splitext(fname)[0])  # 否则用文件名（去除后缀）

# === 按组读取数据并绘图 ===
for i in range(0, len(file_paths), group_size):
    group_files = file_paths[i:i + group_size]   # 取当前组的文件路径列表
    group_labels = labels[i:i + group_size]      # 取当前组的标签列表

    energies = []  # 存放每个文件处理后的能量数组
    spectra = []   # 存放每个文件处理后的谱函数（dN/dE）

    for path in group_files:
        data = sh.getdata(path)  # 读取sdf数据

        # 通过动态属性名选择粒子能量和分布函数
        E_J = getattr(data, f'Grid_allenergy0_{species}').data[0]  # 能量，单位J
        dN = getattr(data, f'dist_fn_allenergy0_{species}').data   # 分布函数

        E_MeV = E_J * E_J_to_MeV  # 转换成MeV

        # 修剪数组长度，保证能被step整除，方便reshape
        length = len(E_MeV)
        if length % step != 0:
            E_MeV = E_MeV[:-(length % step)]
            dN = dN[:-(length % step)]

        # 重新reshape分组合并bins
        E_reshaped = E_MeV.reshape(-1, step)  # shape=(num_bins//step, step)
        N_reshaped = dN.reshape(-1, step)

        # 能量取平均，数量求和，合并成更粗的bin
        E_merged = E_reshaped.mean(axis=1)
        N_merged = N_reshaped.sum(axis=1)

        dE_merged = np.mean(np.diff(E_merged))  # 合并后bin宽度平均值
        spectrum_merged = N_merged / dE_merged  # 计算dN/dE

        energies.append(E_merged)
        spectra.append(spectrum_merged)

    # === 绘图部分 ===
    fig, ax = plt.subplots(figsize=(7, 6))
    for E, S, label in zip(energies, spectra, group_labels):
        ax.semilogy(E, S, label=label, linewidth=2, alpha=0.8)  # 半对数坐标系

    ax.set_xlabel(f'{species} Energy $E$ (MeV)', fontsize=14)  # x轴标签
    ax.set_ylabel(r'd$N$/d$E$', fontsize=14)                   # y轴标签
    ax.set_title(f'Energy Spectra Comparison: Files {i + 1} to {i + len(group_files)}', fontsize=15)  # 标题
    ax.set_xlim(0, 100)  # x轴范围
    ax.tick_params(axis='both', direction='in', which='both', labelsize=12)  # 刻度设置
    ax.legend(fontsize=10)  # 图例
    plt.tight_layout()

    # 如果打开保存开关，保存图片到指定目录
    if save_fig:
        save_path = os.path.join(output_dir, f'spectrum_group_{i // group_size + 1}.png')
        plt.savefig(save_path, dpi=300)
        print(f"Saved figure: {save_path}")

    plt.show()  # 显示图形
```
