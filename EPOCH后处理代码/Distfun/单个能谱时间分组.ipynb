```python
import sdf_helper as sh
import numpy as np
import matplotlib.pyplot as plt
import os

# 参数设置
E_J_to_MeV = 1.0 / 1.6e-13  # J to MeV
step = 2  # 合并bin的大小，比如2代表每2个bin合并成1个bin

# 文件所在目录
base_dir = '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/电子片4'

# 生成文件名列表 distfun0010.sdf ~ distfun0023.sdf
file_nums = range(10, 24)
file_paths = [os.path.join(base_dir, f'distfun{num:04d}.sdf') for num in file_nums]

# 生成标签列表，比如 'distfun0010', 'distfun0011' ...
labels = [f'distfun{num:04d}' for num in file_nums]

# 每组文件数
group_size = 5

# 分组画图
for i in range(0, len(file_paths), group_size):
    group_files = file_paths[i:i+group_size]
    group_labels = labels[i:i+group_size]

    spectra = []
    energies = []

    # 读取数据并合并bin
    for path in group_files:
        data = sh.getdata(path)
        E_J = data.Grid_allenergy0_Photon.data[0]  # 能量 (J)
        dN = data.dist_fn_allenergy0_Photon.data   # 光子数量

        E_MeV = E_J * E_J_to_MeV

        length = len(E_MeV)
        # 如果长度不是step的倍数，去掉多余点保证能成对合并
        if length % step != 0:
            E_MeV = E_MeV[:-(length % step)]
            dN = dN[:-(length % step)]

        # 合并bin
        E_merged = 0.5 * (E_MeV[0::step] + E_MeV[1::step])
        N_merged = dN[0::step] + dN[1::step]

        dE_merged = np.mean(np.diff(E_merged))
        spectrum_merged = N_merged / dE_merged

        energies.append(E_merged)
        spectra.append(spectrum_merged)

    # 画图
    fig, ax = plt.subplots(figsize=(7, 6))
    for E, S, label in zip(energies, spectra, group_labels):
        ax.semilogy(E, S, label=label, linewidth=2, alpha=0.8)

    ax.set_xlabel(r'Photon Energy $E$ (MeV)', fontsize=14)
    ax.set_ylabel(r'd$N$/d$E$', fontsize=14)
    ax.set_title(f'Energy Spectra Comparison: Files {i+1} to {i+len(group_files)}', fontsize=15)
    ax.set_xlim(1, 100)
    ax.tick_params(axis='both', direction='in', which='both', labelsize=12)
    ax.legend(fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
import sdf_helper as sh
import numpy as np
import matplotlib.pyplot as plt
import os

E_J_to_MeV = 1.0 / 1.6e-13
step = 4
# species = "Photon"
species = "Electron"

datasets = [
    {
        'base_dir': '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024+b',
        'file_nums': range(15, 22),  # 范围可以改
        'file_prefix': 'distfun',
        'file_suffix': '.sdf',
        'label_prefix': '片4',
    },
    {
        'base_dir': '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024b4e5',
        'file_nums': range(15, 22),
        'file_prefix': 'distfun',
        'file_suffix': '.sdf',
        'label_prefix': '片6B',
    },
]

# 构造编号->文件路径和标签 映射字典
file_groups = {}

for ds in datasets:
    base_dir = ds['base_dir']
    prefix = ds.get('file_prefix', '')
    suffix = ds.get('file_suffix', '')
    nums = ds['file_nums']
    label_prefix = ds.get('label_prefix', '')

    for num in nums:
        fname = f"{prefix}{num:04d}{suffix}"
        fpath = os.path.join(base_dir, fname)
        label = f"{label_prefix}_{fname}"

        if num not in file_groups:
            file_groups[num] = []
        file_groups[num].append((fpath, label))

# 你只想画编号22的对比图，改成想画的编号列表即可：
numbers_to_plot = [20]

for num in numbers_to_plot:
    if num not in file_groups:
        print(f"编号 {num} 文件不存在，跳过")
        continue

    spectra = []
    energies = []
    labels = []

    for fpath, label in file_groups[num]:
        data = sh.getdata(fpath)
        E_J = getattr(data, f'Grid_allenergy0_{species}').data[0]
        dN = getattr(data, f'dist_fn_allenergy0_{species}').data

        E_MeV = E_J * E_J_to_MeV
        length = len(E_MeV)
        if length % step != 0:
            E_MeV = E_MeV[:-(length % step)]
            dN = dN[:-(length % step)]

        E_merged = 0.5 * (E_MeV[0::step] + E_MeV[1::step])
        N_merged = dN[0::step] + dN[1::step]

        dE_merged = np.mean(np.diff(E_merged))
        spectrum_merged = N_merged / dE_merged

        energies.append(E_merged)
        spectra.append(spectrum_merged)
        labels.append(label)

    # 绘图
    plt.figure(figsize=(8, 6))
    for E, S, lab in zip(energies, spectra, labels):
        plt.semilogy(E, S, label=lab, linewidth=2, alpha=0.8)

    plt.xlabel(r'{} Energy $E$ (MeV)'.format(species), fontsize=14)
    plt.ylabel(r'd$N$/d$E$', fontsize=14)
    plt.title(f'Energy Spectra Comparison: 编号 {num}', fontsize=15)
    # plt.xlim(1, 50)
    plt.tick_params(axis='both', direction='in', which='both', labelsize=12)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.show()
```
