```python
import sdf_helper as sh
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import os

# # 设定数据文件的路径和文件名模式  
base_path = '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024+b'
file_prefix = 'distfun'  
file_suffix = '.sdf'
file_index =  35
file_name = f"{file_prefix}{file_index:04d}{file_suffix}"  

file_path = os.path.join(base_path, file_name)

# 读取数据
distfun_Data = sh.getdata(file_path)
sh.list_variables(distfun_Data)

theta = distfun_Data.Grid_angle_xy_Electron.data[0]*(180/np.pi)
N = distfun_Data.dist_fn_angle_xy_Electron.data

fig,ax = plt.subplots(figsize = (6,6))
ax.plot(theta, N, label="ele",linewidth=2,alpha = 0.7)

# # 设置标签和标题
ax.set_xlabel(r'$E$ (MeV)', fontsize=14)
ax.set_ylabel(r'd$N$/d$E$', fontsize=14)
ax.set_title('Energy Spectrum')
# ax.set_xlim(-40,40)
ax.tick_params(axis='both', direction='in', which='both', labelsize=12)
    # axis='both'：同时设置 x 轴和 y 轴的刻度。
    # direction='in'：将刻度方向设置为向内（指向图形内部）。
    # which='both'：同时设置主刻度和副刻度的刻度方向。
    # labelsize=12：将刻度标签的字体大小设置为 12 磅。
plt.show()
```
