```python
import sdf_helper as sh
import numpy as np
import matplotlib.pyplot as plt
import os

# 设定数据文件的路径和文件名模式
base_path1 = '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024aa'
base_path2 = '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024+b'
file_prefix = 'distfun'
file_suffix = '.sdf'
file_index = 35

# 构造文件路径
file_name = f"{file_prefix}{file_index:04d}{file_suffix}"
file_path1 = os.path.join(base_path1, file_name)
file_path2 = os.path.join(base_path2, file_name)

# 读取第一个文件的数据
distfun_Data1 = sh.getdata(file_path1)
sh.list_variables(distfun_Data1)

px1 = distfun_Data1.Grid_px_py_Electron.data[0]
py1 = distfun_Data1.Grid_px_py_Electron.data[1]
px_py1 = distfun_Data1.dist_fn_px_py_Electron.data

# 对 py 方向进行累加求和
sum_along_py1 = np.sum(px_py1, axis=1)  # axis=1 表示沿着 y 轴（py 方向）求和

# 读取第二个文件的数据
distfun_Data2 = sh.getdata(file_path2)
sh.list_variables(distfun_Data2)

px2 = distfun_Data2.Grid_px_py_Electron.data[0]
py2 = distfun_Data2.Grid_px_py_Electron.data[1]
px_py2 = distfun_Data2.dist_fn_px_py_Electron.data

# 对 py 方向进行累加求和
sum_along_py2 = np.sum(px_py2, axis=1)  # axis=1 表示沿着 y 轴（py 方向）求和

# 创建图像和子图
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制两个文件的累加结果的折线图
ax.plot(px1, sum_along_py1, label=f'Sum along py for file {file_index} (ju2024aa)')
ax.plot(px2, sum_along_py2, label=f'Sum along py for file {file_index} (ju2024+b)')

# 设置标签和标题
ax.set_xlabel(r'$p_x$', fontsize=14)
ax.set_ylabel(r'Sum of $f(p_x, p_y)$ along $p_y$', fontsize=14)
ax.set_title('Comparison of Sum of Distribution Function along $p_y$')
ax.set_xlim(-0.1e-19,0.3e-19)
ax.legend()

plt.show()
```
