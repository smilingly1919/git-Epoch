```python
import os
import numpy as np
import matplotlib.pyplot as plt
import sdf_helper as sh  # 确保你有 sdf_helper.py

# 参数设置
E_J_TO_MeV = 1 / 1.6e-13      # 焦耳转MeV的换算系数
step = 2                    # 合并能量bin大小，比如每2个bin合成1个
species = "Photon"            # 粒子种类，如 "Photon" 或 "Electron"
save_fig = True               # 是否保存图片
output_dir = "/media/lan/Lexar/EPOCH代码/图片/20250701"  # 保存图片的目录

# 数据集信息，包含路径、文件编号、文件名前缀和后缀，以及标签
datasets = [
    {
        'base_dir': '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024a/10kev无磁场',
        'file_nums': [20, 25, 30, 35],
        'file_prefix': 'distfun',
        'file_suffix': '.sdf',
        'custom_labels': ['无磁场-20T0', '无磁场-25T0', '无磁场-30T0', '无磁场-35T0']
    },
    # {
    #     'base_dir': '/media/lan/.../ju2024a',
    #     'file_nums': [20, 25, 30, 35],
    #     'file_prefix': 'distfun',
    #     'file_suffix': '.sdf',
    #     'custom_labels': ['磁场0-18T0', '磁场0-20T0', '磁场0-25T0', '磁场0-30T0']
    # },
    {
        'base_dir': '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch-set2/epoch3d/ju2024ab+2/10kev25磁场',
        'file_nums': [20, 25, 30, 35],
        'file_prefix': 'distfun',
        'file_suffix': '.sdf',
        'custom_labels': ['磁场25-20T0', '磁场25-25T0', '磁场25-30T0', '磁场25-35T0']
    }
]

# 如果需要保存图片，确保目录存在
if save_fig and not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 文件组数（假设每个数据集file_nums长度相同）
n_files = len(datasets[0]['file_nums'])

# 遍历每组文件，做对比绘图
for i in range(n_files):
    file_paths = []
    labels = []

    # 构造每个数据集的完整文件路径和标签
    for ds in datasets:
        num = ds['file_nums'][i]
        fname = f"{ds['file_prefix']}{num:04d}{ds['file_suffix']}"
        fpath = os.path.join(ds['base_dir'], fname)
        file_paths.append(fpath)

        # 先用custom_labels，没有则用文件名
        if 'custom_labels' in ds:
            labels.append(ds['custom_labels'][i])
        else:
            labels.append(fname)

    energies_list = []  # 保存每个文件的合并能量bin中心
    spectra_list = []   # 保存对应的谱函数值

    # 读取每个文件的数据并处理
    for path in file_paths:
        data = sh.getdata(path)  # 读取sdf数据
        E_J = getattr(data, f'Grid_allenergy0_{species}').data[0]  # 能量数组，单位J
        dN = getattr(data, f'dist_fn_allenergy0_{species}').data    # 分布函数数据

        E_MeV = E_J * E_J_TO_MeV  # 转换为MeV

        # 修剪数组长度，保证能被step整除，方便reshape
        trim = len(E_MeV) % step
        if trim != 0:
            E_MeV = E_MeV[:-trim]  # 去掉末尾trim个元素
            dN = dN[:-trim]

        # 重新reshape，合并step个bin
        E_reshaped = E_MeV.reshape(-1, step)      # 每step个一组
        dN_reshaped = dN.reshape(-1, step)

        E_merged = E_reshaped.mean(axis=1)        # 每组bin的能量中心取平均
        N_merged = dN_reshaped.sum(axis=1)        # 每组bin的粒子数求和
        dE_merged = np.mean(np.diff(E_merged))    # bin宽度的平均值

        spectrum = N_merged / dE_merged            # 计算谱密度 dN/dE

        energies_list.append(E_merged)
        spectra_list.append(spectrum)

    # 绘制所有数据集的对比曲线
    plt.figure(figsize=(8, 6))
    for E, S, label in zip(energies_list, spectra_list, labels):
        plt.semilogy(E, S, label=label, linewidth=2)

    plt.xlabel(f'{species} Energy $E$ (MeV)', fontsize=14)
    plt.ylabel('d$N$/d$E$', fontsize=14)
    plt.title(f'Energy Spectrum Comparison - Group {i + 1}', fontsize=15)
    plt.xlim(2, 100)
    plt.legend(fontsize=10)
    plt.grid(True, which='both', linestyle='--', alpha=0.3)
    plt.tight_layout()

    # 保存图片
    if save_fig:
        save_path = os.path.join(output_dir, f"group_{i+1}_spectrum.png")
        plt.savefig(save_path, dpi=300)
        print(f"Saved figure: {save_path}")

    plt.show()
```
