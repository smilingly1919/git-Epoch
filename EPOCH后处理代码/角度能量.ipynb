```python
import sdf_helper as sh
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib.colors import LinearSegmentedColormap

# ========= 1. 文件路径设置 =========
# 设置两个数据文件所在目录（正向锥和倒置锥）
base_path1 = '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch/epoch3d/Cpf-Work/cone/cone1-a=145-1微米'
base_path2 = '/media/lan/4248e56e-6d9a-4026-afad-e8b1d59ceab0/epoch/epoch3d/Cpf-Work/cone/cone1-a=145-1微米倒置'

# 构造文件名
file_prefix = 'distfun'
file_suffix = '.sdf'
file_index = 180
file_name = f"{file_prefix}{file_index:04d}{file_suffix}"

# 完整文件路径
file_path1 = os.path.join(base_path1, file_name)
file_path2 = os.path.join(base_path2, file_name)

# ========= 2. 自定义颜色映射 =========
# 颜色列表从白 → 蓝 → 绿 → 黄 → 红的渐变
color_list = [
    (1.00, 1.00, 1.00),  # 白色
    (0.15, 0.17, 0.50),
    (0.22, 0.32, 0.64),
    (0.24, 0.43, 0.71),
    (0.28, 0.78, 0.92),
    (0.51, 0.79, 0.60),
    (0.74, 0.84, 0.22),
    (0.98, 0.80, 0.07),
    (0.94, 0.37, 0.13),
    (0.92, 0.11, 0.13),
    (0.49, 0.07, 0.08)   # 深红色
]
# 构造线性颜色映射
cmap = LinearSegmentedColormap.from_list("custom_cmap", color_list)

# ========= 3. 数据读取函数 =========
def load_data(file_path):
    """
    从SDF文件中读取 photon 的能谱角度分布函数数据。
    返回：
        - energy_mev: 能量轴，单位 MeV
        - angle_rad: 角度轴，单位 rad
        - dist_fn: 分布函数 dN / (dE * dθ)
    """
    Data = sh.getdata(file_path)
    
    # 获取能量（单位转为 MeV）
    energy_mev = Data.Grid_en_anglexy_Photon.data[0] / 1.6e-13
    
    # 获取角度（单位 rad）
    angle_rad = Data.Grid_en_anglexy_Photon.data[1]
    
    # 计算能量和角度步长，用于归一化
    dE = np.diff(energy_mev).mean()
    dA = np.diff(angle_rad).mean()
    
    # 分布函数归一化为密度
    dist_fn = Data.dist_fn_en_anglexy_Photon.data / (dE * dA)
    
    return energy_mev, angle_rad, dist_fn

# 读取两个数据文件
energy1, angle1, dist1 = load_data(file_path1)
energy2, angle2, dist2 = load_data(file_path2)

# ========= 4. 图形设置 =========
plt.rcParams.update({
    'font.size': 24,          # 主字体大小
    'xtick.labelsize': 18,    # x轴刻度字体
    'ytick.labelsize': 18     # y轴刻度字体
})

# 创建两行一列的极坐标子图
fig, axs = plt.subplots(2, 1, subplot_kw={'projection': 'polar'}, figsize=(9, 13))
fig.subplots_adjust(hspace=0.3)  # 设置子图间距

# ========= 5. 极坐标绘图 =========
# 遍历两个数据集进行绘图
for ax, energy, angle, dist in zip(axs, [energy1, energy2], [angle1, angle2], [dist1, dist2]):
    # 构建极坐标网格
    theta, r = np.meshgrid(angle, energy)
    
    # 绘制伪彩色图
    pcm = ax.pcolormesh(theta, r, dist, shading='auto', cmap=cmap, vmin=0, vmax=1e9)
    
    # 设置径向和角度范围
    ax.set_ylim(0, 200)
    ax.set_xlim(-0.75, 0.75)
    
    # 设置径向刻度
    ax.set_yticks(np.arange(0, 201, 50))
    
    # 不显示默认网格
    ax.grid(False)
    
    # 控制角度标签距离中心的距离
    ax.tick_params(pad=15)

# ========= 6. 颜色条设置 =========
# 添加颜色条，绑定到最后一个 pcolormesh 对象上
cbar = fig.colorbar(pcm, ax=axs, location='right', shrink=0.70, pad=0.05)
cbar.set_label('$dN/(dE \\cdot d\\theta)$', fontsize=24)
cbar.ax.tick_params(labelsize=20)

# 设置指数字体大小和位置
offset_text = cbar.ax.yaxis.offsetText
offset_text.set_fontsize(20)
offset_text.set_position((1.3, 1.19))  # 可根据实际图调整这两个数值

# ========= 7. 显示图像或保存 =========
# plt.savefig("photon_cone_compare.svg", format='svg', bbox_inches='tight')
plt.show()
```
