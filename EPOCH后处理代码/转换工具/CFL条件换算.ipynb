```python
import numpy as np

# 输入参数（单位：米 和 网格数）
Lx = 20e-6  # x方向物理长度
Ly = 10e-6  # y方向物理长度
Lz = 10e-6   # z方向物理长度

Nx = 2000    # x方向网格数
Ny = 1000    # y方向网格数
Nz = 700     # z方向网格数

# CFL 系数（可调）
sigma = 0.8  # 例如取 0.9 表示 90% 安全裕度

# 光速
c = 3.0e8  # m/s

# 分辨率
dx = Lx / Nx
dy = Ly / Ny
dz = Lz / Nz

# CFL 时间步长函数
def cfl_1d(dx, c, sigma):
    return sigma * dx / c

def cfl_2d(dx, dy, c, sigma):
    return sigma / (c * np.sqrt((1/dx**2) + (1/dy**2)))

def cfl_3d(dx, dy, dz, c, sigma):
    return sigma / (c * np.sqrt((1/dx**2) + (1/dy**2) + (1/dz**2)))

# 计算时间步长
dt1d = cfl_1d(dx, c, sigma)
dt2d = cfl_2d(dx, dy, c, sigma)
dt3d = cfl_3d(dx, dy, dz, c, sigma)

# 输出
print(f"CFL系数: σ = {sigma}")
print(f"网格分辨率: dx = {dx:.2e} m, dy = {dy:.2e} m, dz = {dz:.2e} m")
print(f"稳定时间步长:")
print(f"  1D: Δt ≤ {dt1d:.2e} s")
print(f"  2D: Δt ≤ {dt2d:.2e} s")
print(f"  3D: Δt ≤ {dt3d:.2e} s")
```
