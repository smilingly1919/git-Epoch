```python
import math

# 
a0 = 128.0          # 
lambda_um = 0.8   # m

# 
I = (a0 / (8.57e-10 * lambda_um))**2
print(f"I = {I:.2e} W/cm")
```

```python
import math

# 
I = 3.49e+22          # W/cm
lambda_um = 0.8       # m

#  a0
a0 = 8.57e-10 * math.sqrt(I) * lambda_um
print(f"a0 = {a0:.2f}")
```

```python
import math

# 
c = 3e8                    #  (m/s)
epsilon_0 = 8.854e-12      #  (F/m)
e = 1.602e-19              #  (C)
m_e = 9.109e-31            #  (kg)
pi = math.pi

# 
I_W_cm2 = 3.49e+22         #  W/cm
lambda_m = 0.8e-6          #  m

# W/cm  W/m
I_W_m2 = I_W_cm2 * 1e4

# 
omega = 2 * pi * c / lambda_m

#  a0
E0 = math.sqrt(2 * I_W_m2 / (epsilon_0 * c))
a0 = e * E0 / (m_e * c * omega)

print(f"a0 = {a0:.4f}")
```
