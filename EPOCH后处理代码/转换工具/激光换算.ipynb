```python
import math

# 输入
a0 = 128.0          # 归一化振幅
lambda_um = 0.8   # 波长，单位：μm

# 计算强度
I = (a0 / (8.57e-10 * lambda_um))**2
print(f"I = {I:.2e} W/cm²")
```

```python
import math

# 输入
I = 3.49e+22          # 峰值强度，单位：W/cm²
lambda_um = 0.8       # 波长，单位：μm

# 计算 a0
a0 = 8.57e-10 * math.sqrt(I) * lambda_um
print(f"a0 = {a0:.2f}")
```

```python
import math

# 常数
c = 3e8                    # 光速 (m/s)
epsilon_0 = 8.854e-12      # 真空介电常数 (F/m)
e = 1.602e-19              # 元电荷 (C)
m_e = 9.109e-31            # 电子质量 (kg)
pi = math.pi

# 输入
I_W_cm2 = 3.49e+22         # 强度，单位 W/cm²
lambda_m = 0.8e-6          # 波长，单位 m

# 单位转换：W/cm² → W/m²
I_W_m2 = I_W_cm2 * 1e4

# 计算角频率
omega = 2 * pi * c / lambda_m

# 计算 a0
E0 = math.sqrt(2 * I_W_m2 / (epsilon_0 * c))
a0 = e * E0 / (m_e * c * omega)

print(f"a0 = {a0:.4f}")
```
