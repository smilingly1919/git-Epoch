```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

# 设置全局字体为 Times New Roman 和字号
plt.rc('font', family='Times New Roman')
plt.rcParams.update({
    'font.size': 16,                  # 所有字体大小，包括刻度、图例等
    'axes.labelsize': 16,            # x/y 轴标签的字号（xlabel, ylabel）
    'xtick.labelsize': 16,           # x轴刻度字号
    'ytick.labelsize': 16,           # y轴刻度字号
    'legend.fontsize': 16,           # 图例字号
    'axes.titlesize': 16             # 标题字号（title）
})


# 读取 Excel 数据
file_path = r'C:\Users\pengfan\Desktop\第二篇论文\18sdf角动量.xlsx'
df = pd.read_excel(file_path, sheet_name='plot')

# 数据列
x = df['1/e范围']
y1 = df['With B']
y2 = df['Without B']

# 创建图像和坐标轴对象
fig, ax1 = plt.subplots(figsize=(8, 5))

# Plot y1 on ax1
color1 ="#000000"
line1, = ax1.plot(x, y1, color=color1, marker='o', label='With B')
ax1.set_xlabel('1/e Range', fontsize=15)
ax1.set_ylabel(r'$L_\gamma(\mathrm{kg \cdot m^2/s})$', color=color1, fontsize=15)
ax1.tick_params(axis='x', direction='in', labelsize=13)
ax1.tick_params(axis='y', direction='in', labelcolor=color1, labelsize=13)

# 使用科学计数法格式化 y1 轴
ax1.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

# 创建右侧 y 轴
ax2 = ax1.twinx()
color2 = 'tab:red'
line2, = ax2.plot(x, y2, color=color2, marker='s', label='Without B')
ax2.set_ylabel(r'$L_\gamma(\mathrm{kg \cdot m^2/s})$', color=color2, fontsize=16)
ax2.tick_params(axis='y', direction='in', labelcolor=color2, labelsize=13)

# 使用科学计数法格式化 y2 轴
ax2.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax2.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

# 图例
lines = [line1, line2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper right', fontsize=13)

# 标题（可选）
# plt.title('Dual Y-Axis Plot: With B vs Without B', fontsize=15)
# plt.tight_layout()

# 保存为 EPS（可选）
plt.savefig(r'H:\EPOCH代码\图片\20250707\角动量双轴图.eps', format='eps', dpi=600)

plt.show()
```
